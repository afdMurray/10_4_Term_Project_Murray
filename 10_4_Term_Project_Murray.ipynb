```python
# CDF function
def EvalCdf(t, x):
    count = 0.0
    for value in t:
        if value <= x:
            count += 1
            
    prob = count / len(t)
    return prob
    
```

```python
# For runs_scored or X2, the CDF(0.144)
runs = EvalCdf([runs_scored], 0.144)
```
